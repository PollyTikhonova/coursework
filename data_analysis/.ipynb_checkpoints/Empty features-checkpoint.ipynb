```r
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```r
setwd("~/BIO/COURSEWORK/Roytberg/Theme 2/New epoch/correlation")

data = read.csv('../input files/rna-ion-step2/set_minresol_A_7.csv', sep = '\t')
data[is.na(data)] <- 0
#data = data[complete.cases(data), ]
data = data[,2:385]
```

```r
#fold = '../main_datasets/'
fold = '../input files/rna-ion-step2/'
file_ = 'set_minresol_N_5.csv'
```

```r
data = pd.read_csv(fold+file_, sep='\t').fillna(0)
#data.head()
```

```r
np.sum(data['SXVIIm1']>0)
```

```r
import seaborn as sns; 
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
features = list(data.columns)
features.remove('pdb_chain')
data_scaled = min_max_scaler.fit_transform(data[features])
```

```r
k = 0
labels = []
unique_chains = ['']
for chain in data.pdb_chain:
    if (unique_chains[-1] != chain):
        labels.append('%s'%chain)
        unique_chains.append(chain)
    else:
        labels.append('')
```

```r
del data
```

```r
import pickle
with open('data.pickle', 'wb') as f:
     pickle.dump(data_scaled, f)
with open('labels.pickle', 'wb') as f:
     pickle.dump(labels, f)      
```

```r
sns.set()
fig, ax = plt.subplots()
fig.set_size_inches(100, 30)
plt.title('Feature 2003 dataset', fontsize=32);
plt.tick_params(axis='y', which='both', labelleft='off', labelright='on')
sns.heatmap(data_scaled, ax = ax, xticklabels=features, yticklabels='', #labels,
            cbar_kws={"orientation": "horizontal"}, cbar_ax = fig.add_axes([0.15, 0.04, 0.73, 0.02]) );
plt.savefig('outputs/FEATURE_dataset_heatmap.png')
```
