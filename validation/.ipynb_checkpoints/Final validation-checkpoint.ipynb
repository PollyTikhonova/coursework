```python
def load_data(filename):
    '''
        Функция для загрузки датасета без инициализации класса. 
        
        Возвращает:
         data - полный DataFrame
         fetures - список признаков для обучения/предсказания
         x - датасет для тренировки/предсказания в формате np.array
         y - целевой признак, если он есть. В случае с тестовым датасетом возвращает None
    '''
    
    data = pd.read_table(filename, sep='\t')  
    
    data = data[~(data['chainlen']>1000)]
    
    features = list(deepcopy(data.columns))
    features.remove('DSSR')
    
    data.dropna(inplace = True, subset=features)    
    
    [features.remove(column) for column in ['Id','index', 'pdb_chain', 'mg', 'pdb_index'] if column in features];
    x_test = np.array(data[features])
    
    try:
        y_test = np.array(data['mg'])
    except: 
        y_test = None
    change_output('Data loaded')
    return {'data':data, 'features':features, 'x':x_test, 'y':y_test}

def plot_one_plot(plot_elements):
    [eval(plot_string) for plot_string in plot_elements]
```

### Try on 4 sets

```python
folder = '../main_datasets/with_pdb_indexes/'
file_ = "set_minresol-minus-d_A_7_indexes.csv"
m = Magnesium(file_, fold = folder,colours = ['#f6c35b', '#929292'])
```

```python
val2 = load_data(folder+'set_val2_A_7_indexes.csv')
val = load_data(folder+'set_val-minus-minresol_A_7_indexes.csv')
d = load_data(folder+'set_d_A_7_indexes.csv')
```

```python
angles = ['sszp','v1','v2', 'v3', 'v4', 'e-z', 'eta','phase-angle','thetap',
          'SI', 'SII', 'SIII', 'SIV', 'SV','SVII', 'SIX', 'SX', 'SXIII', 'SXIV', 'SXV', 'SXVII', 'SXVIII',
          'SXXI', 'SXXV','SXXVI', 'SXXVII', 'cHS', 'cHH', 'dp']          
angles = np.ravel([[''.join([i,j]) for i in angles] for j in ['m2', 'm1', '', '1', '2']]).tolist()
features = deepcopy(m.features)
[print(i) for i in ['atomN', 'atomOP3', 'chainlen','resol','xray','mg']+angles if i not in features]
[features.remove(i) for i in ['atomN', 'atomOP3', 'chainlen','resol','xray','mg']+angles if i in features];
print('%d features were deleted.' %len(['atomN', 'atomOP3', 'chainlen','resol','xray','mg']+angles))
```

```python
rus = RandomUnderSampler(ratio = 'auto', random_state=42, return_indices=True)
m.trained_model = RandomForestClassifier(n_estimators=200, 
                                                        n_jobs=-1, 
                                                        criterion='gini', 
                                                        max_depth = 26, 
                                                        max_features = 0.7,
                                                        min_samples_leaf = 20)
x = np.array(m.data[features])
y = m.y

x_train, y_train, rus_indexes = rus.fit_sample(x, y)
shuffle_inds = np.arange(x_train.shape[0]).tolist()
#shuffle_inds = shuffle_inds[::-1]
np.random.shuffle(shuffle_inds)
x_train, y_train = x_train[shuffle_inds], y_train[shuffle_inds]
m.trained_model.fit(x_train, y_train)
```

```python
with open('trained_model.pickle', 'wb') as f:
    pickle.dump(m.trained_model, f)
```

```python
predict_train =  m.predict(x = x_train, 
                           y = y_train,
#                              x = np.array(m.data[features]),
#                              y =m.y, 
                             model=m.trained_model)
predict_val2 = m.predict(x = np.array(val2['data'][features]),
                             y = val2['y'], 
                             model=m.trained_model)
predict_val = m.predict(x = np.array(val['data'][features]),
                             y = val['y'], 
                             model=m.trained_model)
predict_d = m.predict(x = np.array(d['data'][features]),
                             y = d['y'], 
                             model=m.trained_model)
```

```python
plt.rcParams['figure.facecolor'] = 'white'

fig = plt.figure(figsize = (20, 12))
fig.set_facecolor('white')
fig.suptitle('RFC. MaxDepth=26, maxFeatures = 0.7. Min samples leaf = 20. Shuffle Train. Without duplicate features. -phase, thetap, -eta', fontsize=18, fontweight='bold')
plt.subplots_adjust(hspace=0.3, wspace=0.1)

fig.add_subplot(2,2, 1)
plot = predict_train['plots']['prob_density']
[eval(plot_string) for plot_string in plot]
plt.title('Train. Probabilty Distribution. \n Portion of sites: %.3f. Number of atoms: %d'%
          (np.sum(y_train==1)/y_train.shape[0], y_train.shape[0]))
fig.add_subplot(2,2, 2)
plot = predict_d['plots']['prob_density']
[eval(plot_string) for plot_string in plot]
plt.title('Set d A7. Probabilty Distribution. \n Portion of sites: %.3f. Number of atoms: %d'%
          (np.sum(d['y']==1)/d['y'].shape[0], d['y'].shape[0]));
fig.add_subplot(2,2, 3)
plot = predict_val2['plots']['prob_density']
[eval(plot_string) for plot_string in plot]
plt.title('Set val2 A7. Probabilty Distribution. \n Portion of sites: %.3f. Number of atoms: %d'%
          (np.sum(val2['y']==1)/val2['y'].shape[0], val2['y'].shape[0]))

fig.add_subplot(2,2, 4)
plot = predict_val['plots']['prob_density']
[eval(plot_string) for plot_string in plot]
plt.title('Set val minus A7 (minresol). Probabilty Distribution. \n Portion of sites: %.3f. Number of atoms: %d'%
          (np.sum(val['y']==1)/val['y'].shape[0], val['y'].shape[0]));
display(HTML('<b>Prediction scores</b>'))
print('{:30} {:.3f}'.format('Set test (minresol):', predict_train['test_score']))
print('{:30} {:.3f}'.format('Set val2 A7:', predict_val2['test_score']))
print('{:30} {:.3f}'.format('Set val - minus minresol A7:', predict_val['test_score']))
print('{:30} {:.3f}'.format('Set d A7:', predict_d['test_score']))
```

### 1hc8, 1qa6

```python
folder = '../main_datasets/with_pdb_indexes/'
file_ = "set_minresol-minus-d_A_7_indexes.csv"
m = Magnesium(file_, fold = folder,colours = ['#f6c35b', '#929292'])
```

```python
hc = m.data.loc[np.array([i.split('.cif1_')[0] for i in m.data.pdb_chain]) == '1hc8']
m.data = m.data.loc[np.array([i.split('.cif1_')[0] for i in m.data.pdb_chain]) != '1hc8']
```

```python
hcqa = load_data('../input files/1qa6.csv')['data']
```

```python
hcqa['pdb_index'] = [i.split('_')[1] for i in hcqa.DSSR]
hcqa = hcqa.append(hc)
del hc
```

```python
angles = ['sszp','v1','v2', 'v3', 'v4', 'e-z', 'eta','phase-angle','thetap',
          'SI', 'SII', 'SIII', 'SIV', 'SV','SVII', 'SIX', 'SX', 'SXIII', 'SXIV', 'SXV', 'SXVII', 'SXVIII',
          'SXXI', 'SXXV','SXXVI', 'SXXVII', 'cHS', 'cHH', 'dp']          
angles = np.ravel([[''.join([i,j]) for i in angles] for j in ['m2', 'm1', '', '1', '2']]).tolist()
features = deepcopy(m.features)
[print(i) for i in ['atomN', 'atomOP3', 'chainlen','resol','xray','mg']+angles if i not in features]
[features.remove(i) for i in ['atomN', 'atomOP3', 'chainlen','resol','xray','mg']+angles if i in features];
print('%d features were deleted.' %len(['atomN', 'atomOP3', 'chainlen','resol','xray','mg']+angles))
```

```python
m.x = m.data[m.features]
var = np.var(m.x, axis = 0)
features = deepcopy(m.features)
k = len(features)
features = list(var[var > 0.001].axes[0])
k -= len(features)
print(len(features))
angles = ['sszp','v1','v2', 'v3', 'v4', 'e-z', 'eta','phase-angle','thetap', 'suite']          
angles = np.ravel([[''.join([i,j]) for i in angles] for j in ['m2', 'm1', '', '1', '2']]).tolist()    
[print(i) for i in ['mg']+angles if i not in features]
[features.remove(i) for i in ['mg']+angles if i in features];
print('%d features were deleted.' %(len(['mg']+angles)+k))
```

```python
rus = RandomUnderSampler(ratio = 'auto', random_state=42, return_indices=True)
m.trained_model = RandomForestClassifier(n_estimators=200, 
                                                        n_jobs=-1, 
                                                        criterion='gini', 
                                                        max_depth = 26, 
                                                        max_features = 0.7,
                                                        min_samples_leaf = 20)
x = np.array(m.data[features])
y = m.data['mg']

x_train, y_train, rus_indexes = rus.fit_sample(x, y)
shuffle_inds = np.arange(x_train.shape[0]).tolist()
#shuffle_inds = shuffle_inds[::-1]
np.random.shuffle(shuffle_inds)
x_train, y_train = x_train[shuffle_inds], y_train[shuffle_inds]
m.trained_model.fit(x_train, y_train)
```

```python
with open('trained_results_validation/1hc8, 1qa6 remove on var 001/trained_model.pickle', 'wb') as f:
    pickle.dump(m.trained_model, f)
```

```python
predict_hcqa = m.predict(x = hcqa[features],
                             y = hcqa['mg'], 
                             model=m.trained_model)
```

```python
full_pdbs = np.array([i.split('.')[0] for i in hcqa['pdb_chain']])
pdbs = np.unique(full_pdbs)
plot_train = {}
for pdb in pdbs:
    ind = pdb==full_pdbs
    if np.sum(predict_hcqa['y'].values[ind] == 1) > 1:
        y, y_prob, y_pred = predict_hcqa['y'][ind], predict_hcqa['probability'][ind], predict_hcqa['prediction'][ind]
        plot_train[pdb] = plot_probability_density(y_prob, y)
        plot_train[pdb].append(form_plot_string('plt.title', "%s. Score %.2f. Atoms - %d. MG - %d"% 
                                                (pdb, f1_score(y, y_pred), y.shape[0], np.sum(y==1))))
    else:
        print(pdb)
```

```python
plt.rcParams['figure.facecolor'] = 'white'
fig = plt.figure(figsize=(20, 10))
for i, program in tqdm_notebook(enumerate(plot_train.keys())):
    plt.subplots_adjust(top = 0.98, bottom = 0.05,  hspace=0.3, wspace=0.1)
    fig.add_subplot(len(plot_train.keys())//2+1, 2, i+1)
    [eval(plot_string) for plot_string in plot_train[program]]
```

```python
hcqa['probability'] = predict_hcqa['probability']
```
