```python
import sys
sys.path.insert(0, '../basic codes')

import class_magnesium_kaggle #_transform_pairings
from class_magnesium_kaggle import *

from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedShuffleSplit
```

### Train A_5

```python
rfc = RandomForestClassifier(n_estimators = 100, n_jobs=-1)
```

```python
filename = ["set_minresol_N_5.csv", "set_minresol_M_5.csv", "set_minresol_A_5.csv", 
            "set_minresol_N_7.csv", "set_minresol_M_7.csv", "set_minresol_A_7.csv"]
fold = '../input files/rna-ion-step2/'
```

```python
m = Magnesium(filename[0], fold = fold)
```

```python
skf = StratifiedShuffleSplit(n_splits=1)
rus = RandomUnderSampler(ratio = 'auto', random_state=42)
```

```python
results = {}
for file in tqdm_notebook(filename, desc='Files', leave=True):
    print(file)
    data = pd.read_csv(fold+file, sep='\t').dropna()
    features = list(data.columns)
    features.remove('mg')
    features.remove('pdb_chain')
    x = data[features].values
    y = data['mg'].values
    score = []
    print('Data is loaded.')

    for train_index, test_index in skf.split(x, y):  
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]
        x_train, y_train = rus.fit_sample(x_train, y_train)
      #  x_test, y_test = rus.fit_sample(x_test, y_test)
        rfc.fit(x_train, y_train)
        results[file] = m.predict(x=x_test, y=y_test, model=rfc)
        results[file]['accuracy'] = accuracy_score(results[file]['y'], results[file]['prediction'])
```

```python
results_balanced = {}
for file in tqdm_notebook(filename, desc='Files', leave=True):
    print(file)
    data = pd.read_csv(fold+file, sep='\t').dropna()
    features = list(data.columns)
    features.remove('mg')
    features.remove('pdb_chain')
    x = data[features].values
    y = data['mg'].values
    score = []
    print('Data is loaded.')

    for train_index, test_index in skf.split(x, y):  
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]
        x_train, y_train = rus.fit_sample(x_train, y_train)
        x_test, y_test = rus.fit_sample(x_test, y_test)
        rfc.fit(x_train, y_train)
        results_balanced[file] = m.predict(x=x_test, y=y_test, model=rfc)
        results_balanced[file]['accuracy'] = accuracy_score(results_balanced[file]['y'], results_balanced[file]['prediction'])
```
