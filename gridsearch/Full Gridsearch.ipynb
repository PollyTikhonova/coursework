```python
def gridsearch_function(filename, fold, n_splits, parametres):
    m = Magnesium(filename, fold = fold,colours = ['#f6c35b', '#929292'])
    m.groups = np.array(m.groups)[m.data['chainlen'].values<1000]
    m.data = m.data[~(m.data['chainlen']>=1000)]    
    m.x = m.data[m.features]
    m.y = m.data['mg']
    var = np.var(m.x, axis = 0)
    features = deepcopy(m.features)
    features = list(var[var > 0.015].axes[0])
    print(len(features))
    [features.remove(i) for i in ['mg'] if i in features];
    to_del = [i for i in np.unique(m.groups) if np.sum(m.y[i==np.array(m.groups)]==1) == 0]
    ind = [i==np.array(m.groups) for i in to_del]
    m.data = m.data.iloc[np.where(np.sum(ind, axis=0)==0)]
    m.groups = np.array(m.groups)[np.where(np.sum(ind, axis=0)==0)]
    m.x = m.data[features].values
    m.y = m.data['mg'].values
    
    rfc = RandomForestClassifier(n_estimators=200, n_jobs=-1)
    rus = RandomUnderSampler(ratio = 'auto', random_state=42, return_indices = True)
    gss = GroupShuffleSplit(n_splits=n_splits, test_size=0.3)
    
    splittings = []
    for train_index, test_index in gss.split(m.x, m.y, groups = m.groups):
        x_train, y_train, train_indices = rus.fit_sample(m.x[train_index], m.y[train_index])
        x_test, y_test, test_indices = rus.fit_sample(m.x[test_index], m.y[test_index])
#         shuffle_indices = np.arange(x_train.shape[0]).tolist()
#         np.random.shuffle(shuffle_indices)
#         splittings.append([train_index[shuffle_indices], test_index[test_indices]])
        splittings.append([train_index[train_indices], test_index[test_indices]])
    if not os.path.isdir('outputs'):
        os.mkdir('outputs')
    with open("outputs/splitting_indices.sav", 'wb') as file_to_save:
            pickle.dump(splittings, file_to_save)

#     m = Magnesium(filename, fold = fold,colours = ['#f6c35b', '#929292'])
#     if 'group' in m.features:
#         m.features.remove('group')
# #     m.data = m.data[~(m.data['chainlen']>1000)]
#     m.x = m.data[m.features].values
#     m.y = m.data.mg.values
    
#     rfc = RandomForestClassifier(n_estimators=200, n_jobs=-1)
#     rus = RandomUnderSampler(ratio = 'auto', random_state=42, return_indices = True)
    
#     splittings = []
#     gss = GroupShuffleSplit(n_splits = n_splits, test_size = 0.3, random_state = 0).split(m.x, m.y,groups = m.groups)
#     for train_index, test_index in gss:
#         x_train, y_train, train_indices = rus.fit_sample(m.x[train_index], m.y[train_index])
#         x_test, y_test, test_indices = rus.fit_sample(m.x[test_index], m.y[test_index])
# #         shuffle_indices = np.arange(x_train.shape[0]).tolist()
# #         np.random.shuffle(shuffle_indices)
#         splittings.append([train_index[train_indices], test_index[test_indices]])
#     with open("outputs/splitting_indices.sav", 'wb') as file_to_save:
#             pickle.dump(splittings, file_to_save)
    
    grid = GridSearchCV(rfc, verbose = 17,
                       cv=splittings,
                       param_grid=parametres, 
                        scoring={'precision':precision_scorer,
                                                       'recall':recall_scorer,
                                                       'f1':f1_scorer, 
                                                       'accuracy':accuracy_scorer},
                        refit=False, return_train_score=True,
                       n_jobs = -1)
    grid.fit(m.x, m.y)
    data = pd.DataFrame(grid.cv_results_)
    
    # save_data
#    dataframe = pd.DataFrame(data)[[2,3,4,5, -2, -1]]        
    data.to_csv('outputs/'+'%s_grid_data.csv'%filename.split('.csv')[0], index = False, sep = '\t')
    try:
        return [pd.DataFrame(data), grid.best_params_, grid.best_score_, grid.best_estimator_]
    except AttributeError:
        return [pd.DataFrame(data)]
```

```python
import time
def get_time(start_time):
    t_sec = round(time.time() - start_time)
    (t_min, t_sec) = divmod(t_sec,60)
    (t_hour,t_min) = divmod(t_min,60) 
    return 'Time passed: {} hour {} min {} sec'.format(t_hour,t_min,t_sec)
```
